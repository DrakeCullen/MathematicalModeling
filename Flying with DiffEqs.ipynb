## Description
- We are looking into the situation that involved mechanical resonance and caused two plane crashes in 1959 and 1960
- Initially, the planes start at rest and the wings are at equilibrium
- As the propellers on the wings begin to spin, they cause an external force on the wings
- The external force displaces the wings in the upward and downward directions
- The planes wing-body joint acts as a spring and applied more force on the wing the further it is from equilibrium
- If the wings are unable to withstand the force, they will snap

## Given:
- We will evaluate the fluttering motion at the wings center of masss
- Think of the wing as a spring mass system
- The wing has a mass of m kg
- The wing body joint acts like a spring with a spring constant of k N/m
- The center of mass moves up and down in a straight line
- Damping forces are negligible
- The wing starts at rest
- At time t = 0, the propeller begins the vibrate with a force equal to $$ F(t)=F_{0}sin(wt) $$

## Background Knowledge:
$$ F=ma $$ $$ F=-ky $$ $$ F=\frac{d^2y}{dt^2}$$



## Part 1) A,B,C
Find the equation of motion for the wing's center-of-mass. Be sure to include the appropriate initial conditions you would need to solve the differential equation. Given values, solve the equation.
$$
m = 850 kg \\
k = 13,600 N/m \\
w = 8 \\
F_0 = 1,550 N
$$

From our background knowledge, we know: $$\frac{d^2y}{dt^2}+ky=0$$
By plugging our given values in and knowing that the force must be equal to the external force, we get: 
$$ 850\frac{d^2y}{dt^2}+13600y=1550sin(8t)$$
This becomes: $$ 850y^2+13600y=1550sin(8t) $$

```python
# Import the required modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.pyplot import figure
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from matplotlib.animation import FuncAnimation
# This makes the plots appear inside the notebook
%matplotlib inline
```

```python
# 100 linearly spaced numbers
x = np.linspace(0,10,100)

# the function, which is y = sin(x) here
y = 31/408*np.sin(4*x)-31/816*np.sin(8*x)

# setting the axes at the centre
fig = plt.figure()
fig.set_size_inches(14.5, 8.5)
plt.rcParams.update({'font.size': 20})  # increase the font size
ax = fig.add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_label_coords(1,.47)
plt.ylabel("Displacement")
plt.xlabel("Time")
plt.title("Displacement vs Time for Airplane Wing")


# plot the function
plt.plot(x,y, 'b-')

# show the plot
plt.show()
```

```python
def f(u,x):
    return (850*u[1],-13600*u[1]+1550*np.sin(8*x))

y0 = [0,0]
xs = np.linspace(1,10,200)
us = odeint(f,y0,xs)
ys = us[:,0]/100
plt.plot(xs,ys)
plt.rcParams.update({'font.size': 15})
plt.title("Wing Displacement Vs Time")
plt.ylabel("Displacement")
plt.xlabel("Time")
plt.show()
```
